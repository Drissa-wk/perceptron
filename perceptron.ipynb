# 1. Import perceptron

```python
from perceptron import Perceptron
```

# 2. Use the perceptron with random data

## a. create random data with the make_blobs function of scikit-learn

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=100, centers=2, n_features=2, random_state=0)
```

## b. Initialize a perceptron object

```python
perceptron = Perceptron()
```

## c. train the model with our data

```python
perceptron.train(X, y)
```

## d. make predictions with the trained model and evaluate the accuracy

```python
y_pred = perceptron.predict(X)
print(perceptron.accuracy(y, y_pred))
```

## e. draw the loss function curve

```python
perceptron.draw_loss()
```

## f. draw the classification

> **Note** : this function provides a funny animation that shows the evolution of the classification (during the training phase) after each iteration.
It's only available for data with **two(02)** features !

```python
perceptron.draw_classification(X, y)
```

# 2. Work with real data

> **Note** : We'll use the **Framingham Heart Study** dataset used in cardiovascular research to identify common factors and characteristics that contribute to the development of cardiovacular disease.
It's available in the **dataset** folder.

## a. imports

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

## b. load data

```python
df = pd.read_csv("dataset/framingham.csv")
```

## c. preprocessing

> **Note** : After a preprocessing we end up with the conclusion that the columns **'currentSmoker', 'TenYearCHD', 'heartRate'** are not useful in this analysis so we drop these three columns and split our features (**X**) and our target (**y**)

```python
X = df.drop(['currentSmoker', 'TenYearCHD', 'heartRate'], axis=1)
Y = df['TenYearCHD']
```

## d. split in train and test sets

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
```

## e. initialize and train a model

```python
perceptron2 = Perceptron()
perceptron2.train(X_train, Y_train)
```

## f. test and accuracy

```python
y_pred = perceptron2.predict(X_test)
print(perceptron2.accuracy(Y_test, y_pred))
```

## g. loss function

```python
perceptron.draw_loss()
```

## h.draw classification

```python
perceptron.draw_classification(X_test, Y_test)
```
